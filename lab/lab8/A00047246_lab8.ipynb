```python
import statsmodels.api as sm
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_st = pd.read_csv('/content/drive/MyDrive/TUD/Na Li/lab10/student_math_clean.csv')
```

```python
data_st.head()
```

```python
X1 = data_st['study_time']
y1 = data_st['final_grade']
```

```python
X1 = sm.add_constant(X1)
```

```python
study_time_mapping = {
    '<2 hours': 1,
    '2 to 5 hours': 2,
    '5 to 10 hours': 3,
    '>10 hours': 4
}

X1_numeric = data_st['study_time'].map(study_time_mapping)
X1_numeric = sm.add_constant(X1_numeric)

model1 = sm.OLS(y1, X1_numeric).fit()
print(model1.summary())
```

```python
X2 = data_st['absences']
y2 = data_st['final_grade']

X2 = sm.add_constant(X2)

model2 = sm.OLS(y2, X2).fit()
print(model2.summary())
```

```python
X3 = data_st['class_failures']
y3 = data_st['final_grade']

X3 = sm.add_constant(X3)

model3 = sm.OLS(y3, X3).fit()
print(model3.summary())
```

```python
X4 = data_st[['study_time', 'absences', 'class_failures']]
y4 = data_st['final_grade']

# Convert 'study_time' to numerical using the existing mapping
study_time_mapping = {
    '<2 hours': 1,
    '2 to 5 hours': 2,
    '5 to 10 hours': 3,
    '>10 hours': 4
}
X4['study_time'] = X4['study_time'].map(study_time_mapping)

X4 = sm.add_constant(X4)

model4 = sm.OLS(y4, X4).fit()
print(model4.summary())
```
